- **block_num** = Nmero do bloco atual. Quando o tesseract faz o OCR, ele divide a imagem em vrias regies, o que pode variar de acordo com os parametros do PSM e tambm outros critrios prprios do algoritmo. Cada bloco  uma regio

- **conf** = confiana da predio (de 0 a 100. -1 significa que no foi reconhecido texto)

- **height** = altura do bloco de texto detectada (ou seja, da caixa delimitadora)

- **left** = coordenada x onde inicia a caixa delimitadora

- **level** = o level (nvel) corresponde  categoria do bloco detectado. so 5 valores possiveis:
  1. pgina
  2. bloco
  3. pargrafo
  4. linha
  5. palavra

Portanto, se foi retornado o valor 5 significa que o bloco detectado  texto, se foi 4 significa que o que foi detectado  uma linha

- **line_num** = nmero da linha do que foi detectado (inicia com 0)

- **page_num** = o ndice da pgina onde o item foi detectado. Na maioria dos casos sempre haver uma pgina s

- **text** = o resultado do reconhecimento

- **top** = coordenada y onde a caixa delimitadora comea

- **width** = largura do bloco de texto atual detectado

- **word_num** = numero da palavra (indice) dentro do bloco atual

## Caixa delimitadora

````python
def caixa_texto(resultado, img, cor = (255, 100, 0)):
  x = resultado['left'][i]
  y = resultado['top'][i]
  w = resultado['width'][i]
  h = resultado['height'][i]

  cv2.rectangle(img, (x, y), (x+w, y+h), cor, 2)

  return x, y, img
````

## Funo para fontes

````python
def escreve_texto(texto, x, y, img, fonte, tamanho_texto=32):
  fonte = ImageFont.truetype(fonte, tamanho_texto)
  img_pil = Image.fromarray(img)
  draw = ImageDraw.Draw(img_pil)
  draw.text((x, y - tamanho_texto), texto, font = fonte)
  img = np.array(img_pil)
  return img
````

````python
img_copia = rgb.copy()
for i in range(len(resultado['text'])):
  confianca = int(resultado['conf'][i])
  if confianca > min_conf.value:
    x, y, img = caixa_texto(resultado, img_copia)

    texto = resultado['text'][i]
    #cv2.putText(img_copia, texto, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,100,255))
    img_copia = escreve_texto(texto, x, y, img_copia, fonte)
      
plt.imshow(img_copia)
````

## Regular Expressions - RE

Nas aulas anteriores, ns aprendemos a detectar palavras, colocar *bounding boxes* e a adicionar textos que o OCR tambm estava identificando. Agora, vamos aprender a retirar/capturar textos especficos.

Considerando estes casos, ns usaremos a biblioteca **RE** ou **Regular Expressions**. Ela nos ajudar a encontrar as expresses regulares, como e-mails ou datas, enfim, elementos que aparecem com muita frequncia. No [site da RE, disponvel neste link](https://regexr.com/), ns encontramos algumas dessas expresses frequentes. Alis, elas tm um formato muito semelhante, o que facilita a nossa busca. No site da expresso regular ou regex, existe um padro de dados especficos, que eu copie, salvei e que ns usaremos no cdigo. Ento, escreveremos `padrao_data` igual ao cdigo do regex que retiramos do site. 

```
padrao_data = '^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|1[012])/(19|20)\d\d$'

```

Esse  o padro de data que usaremos para fazer a seguinte verificao: se esse cdigo funcionar, isto , for igual ao padro de data que j est dentro do nosso texto, podemos prosseguir e adicionar texto. Agora que j colocamos o padro de data, podemos pegar a nossa funo anterior e tornar o padro de data uma nova funo. 

```
img_copia = rgb.copy()
for i in range(0, len(resultado['text'])):
  confianca = int(resultado['conf'][i])
  if confianca > min_conf:
    texto = resultado['text'][i]

    if re.match(padrao_data, texto):
      x, y, img = caixa_texto(resultado, img_copia)
      img_copia = escreve_texto(texto, x, y, img_copia, fonte, 12)
    else:
      x, y, img_copia = caixa_texto(resultado, img_copia)

cv2_imshow(img_copia)

```

Comeamos fazendo a cpia da nossa imagem, `img_copia`. Depois, fazemos o `for` de zero at o resultado, o nmero de textos. Em seguida, ele vai conferir a `confianca`. Se ela for igual ao valor mnimo, `confianca > min_conf`, que  40, ele vai passar para a prxima linha: `texto = resultado['text'][i]`. Essa parte ns j havamos feito.

A segunda parte  nova: se dentro da biblioteca que importamos, que  a `re`, o padro de data for igual ao texto, faremos tanto a caixa de texto quanto a escrita do texto; se o padro no for igual, faremos apenas a caixa de texto e no escreveremos nada. Por fim, mostraremos a imagem. Ele est fazendo um if/else dentro do nosso prprio `if`. 

O `texto` ficou antes do `caixa_texto` para que ele pudesse participar do nosso `if` que est mais abaixo. Vamos rodar e conferir o resultado.

````python
import re
````

````python
img = cv2.imread('./text-recognize/Imagens/Aula4-tabela_teste.png')
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
````

````python
config_tesseract = "--tessdata-dir ./exercicios/tessdata"
resultado = pytesseract.image_to_data(rgb, config=config_tesseract, lang="por", output_type=Output.DICT)
resultado
````

````python
padrao_data = '^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|1[012])/(19|20)\d\d$'
````

````python
img_copia = rgb.copy()
for i in range(0, len(resultado['text'])):
  confianca = int(resultado['conf'][i])
  if confianca > 40:
    texto = resultado['text'][i]

    if re.match(padrao_data, texto):
      x, y, img = caixa_texto(resultado, img_copia)
      img_copia = escreve_texto(texto, x, y, img_copia, fonte, 12)
    else:
      x, y, img_copia = caixa_texto(resultado, img_copia)

plt.imshow(img_copia)
````

Ns temos um caractere "E" na frase principal da xcara, "VAI. E SE DER SONO". Este "E" sumiria, portanto,  um caractere a menos em relao a dois falsos positivos. Precisamos ponderar qual  a melhor opo: ter dois falsos positivos ou perder uma visualizao de letra. Eu considero essa perda aceitvel para eliminarmos os dois falsos positivos, j que no vamos trabalhar com um pr-processamento da imagem.


Comeamos fazendo a cpia da imagem, depois fizemos o `for`, onde pegamos a `confianca`. Se a confiana for maior que o mnimo da confiana que estipulamos no nosso projeto, 40, o texto ser igual ao texto de resultado. 

Depois, se no existir nada no espao, se tivermos apenas um caractere ou menos que um, ele no adicionar a caixa de texto. Por consequncia, se existir mais que um caractere, ele adicionar caixa de texto. Essa lgica pode parecer um pouco confusa no comeo, mas, quando partirmos para a prtica.
